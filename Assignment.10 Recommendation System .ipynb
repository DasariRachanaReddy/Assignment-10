```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("book.csv", encoding='Unicode_escape')
```

```python
df
```

```python
#number of unique users in the dataset
df.nunique()
```

```python
df.drop(df.columns[[0]],axis=1,inplace =True)
df
```

```python
#Renaming the colums
df.columns = ["UserID","BookTitle","BookRating"]
```

```python
df
```

```python
len(df.UserID.unique())
```

```python
##user item collabrative
book_df = df.pivot_table(index='UserID',
                   columns='BookTitle',
                   values='BookRating').reset_index(drop=True)
```

```python
book_df
```

```python
book_df.fillna(0,inplace=True)
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book = 1 - pairwise_distances( book_df.values,metric='cosine')
```

```python
book
```

```python
#Store the results in a dataframe
book_s = pd.DataFrame(book)
```

```python
#Set the index and column names to user ids 
book_s.index = df.UserID.unique()
book_s.columns = df.UserID.unique()
```

```python
book_s.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(book, 0)
book_s.iloc[0:5, 0:5]
```

```python
#Most Similar Users
print(book_s.idxmax(axis=1)[1348])
print(book_s.max(axis=1).sort_values(ascending=False).head(50))
```
